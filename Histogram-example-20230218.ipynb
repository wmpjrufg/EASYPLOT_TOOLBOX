```python
# Data
DF =  pd.DataFrame({'col1': np.random.normal(0, 1, 1000),
                     'col2': np.random.normal(5, 2, 1000),
                     'col3': np.random.normal(-5, 3, 1000)})
COLUMN = 'col1'
NAMES = np.array([['$x_{1}$', '#8C0C15'], 
                  ['$x_{2}$', '#5FD34D'],
                  ['$x_{3}$', '#4DA7D3']])

ID = 0

# Plot in looping
for COLUMN in DF:
    PLOT_SETUP = {
              'NAME': f"{COLUMN}_Histogram",
              'EXTENSION': 'svg',
              'WIDTH': 0.20, 
              'HEIGHT': 0.10,
              'X AXIS LABEL': NAMES[ID, 0],
              'X AXIS SIZE': 15,
              'Y AXIS LABEL': 'Frequency',
              'Y AXIS SIZE': 15,
              'AXISES COLOR': '#000000',
              'LABELS SIZE': 15,
              'LABELS COLOR': '#000000', 
              'CHART COLOR': NAMES[ID, 1],
              'KDE': True,
              'BINS': 20,
              'DPI': 600,
             }
    
    DATA = {
            'DATASET': DF,
            'COLUMN': COLUMN        
           }
    HISTOGRAM_CHART(DATA, PLOT_SETUP)
    ID += 1
```
